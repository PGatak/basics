```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits.mplot3d as m3d
```

```python
x = np.linspace(-10,10,6)
y = x**2
```

```python
plt.figure(1)
plt.title("Przyklad krzywej lamanej")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, color = "k")
```

```python
x = np.linspace(-10,10,200)         
y = x**2
plt.figure(2)
plt.title("Przyklad krzywej ciaglej")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y , color = "r")
```

```python
flights = sns.load_dataset("flights")
plt.figure(3)
plt.title("Dane lotnicze")
plt.ylabel("Liczba pasazerow")
flights["passengers"].plot(color = "y")
```

```python
plt.figure(4)
plt.title("Dane lotnicze")
plt.ylabel("Srednia roczna liczba pasazerow")
flights["passengers"].groupby(flights["year"]).agg(np.mean).plot(color = (0.9,0.7,0.2))
```

```python
plt.figure(5)
plt.title("Dane lotnicze - wykres rozproszeniowy")
plt.scatter(flights.year,flights.passengers, s = 10, marker = "2" , color = "g")
```

```python
t = np.linspace(0,2*np.pi,100,endpoint=False)
x = 16*np.sin(t)**3
y = 16*np.cos(t)
plt.figure(6)
plt.title("Figura z wypelnieniem")
plt.text(9,14,r"$f(t)=16sin^3t$")
plt.fill(x , y , fill = (0.1,0.2,0.4) , hatch = "*", alpha = 0.5)
```

```python
x = np.arange(0,10,0.1)
y = x**3
z = -(x**3)
w = 0*x
plt.figure(7)
plt.title("Rozne rodzaje linii")
plt.xlabel("x")
plt.ylabel("y,z,w")
plt.plot(x,y, "--", label = "$x^3$")
plt.plot(x,z, "-.", label = "$-x^3$")
plt.plot(x,w, ":", label = "$0$")
plt.legend()
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = x**3
plt.figure(8)
plt.title("Szybsze formatowanie")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, "r.")
```

```python
x = np.arange(0,10,0.1)
y = x+1
plt.figure(9)
plt.title("Formtowanie osi")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, "r-.")
plt.axis("off")
plt.axis("equal")
plt.axis("tight")
plt.axis([-1,10,0,12])
```

```python
x = np.r_[1,2,3]
y = np.r_[2,1,7]
plt.figure(10)
plt.title("Etykiety punktow")
plt.scatter(x,y)
plt.xticks([1,2,3],["A","B","C"])
plt.yticks([2,1,7],["A'","B'","C'"])
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.figure(11)
plt.subplot(2,1,1)
plt.title("Podzial wykresu")
plt.plot(x,y, color = "r")
plt.subplot(2,1,2)
plt.title("Podzial wykresu")
plt.plot(x,y , color = "g")
```

```python
procenty = [0.5, 0.4, 0.1]
odp = ["Tak","Nie","Nie wiem"]

plt.figure(12)
plt.subplot(1,2,1)
x = np.arange(len(procenty))
s = 0.5
plt.bar(x,procenty,s,color = ["r","g","k"])
plt.xticks(x,odp, rotation = 0)
plt.subplot(1,2,2)
plt.pie(procenty,labels = odp, colors = ["r","g","k"])
plt.axis("equal")
```

```python
f_p = flights.pivot("month","year","passengers")
sns.heatmap(f_p,annot = True,fmt = "d", linewidths = 0.5)
```

```python
tips = sns.load_dataset("tips")
tips["tip_frac"]=tips["tip"]/tips["total_bill"]
sns.boxplot(tips["tip_frac"],tips["day"],tips["sex"],palette = ["0.2","0.8"])
```

```python
plt.figure(13)
plt.subplot(1,2,1)
a1 = sns.distplot(tips.tip_frac[tips.sex == "Male"], vertical = False)
plt.title("Male")
plt.subplot(1,2,2, sharey = a1)
a2 = sns.distplot(tips.tip_frac[tips.sex == "Female"], vertical = False)
plt.title("Female")
```

```python
plt.figure(14)
plt.hist(tips.tip_frac,bins=10)
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris,diag_kind="kde",vars = ["sepal_length","sepal_width","petal_length"])
```

```python
x = y = np.linspace(-10,10,100)
X,Y = np.meshgrid(x,y)
Z = X**2 - Y**2

plt.figure(15)
plt.contour(X,Y,Z)
```

```python
plt.contourf(X,Y,Z)
```

```python
plt.figure(17)
fig = plt.figure()
sns.set(rc={"axes.facecolor":"white","figure.facecolor":"white"})
ax = m3d.Axes3D(fig,azim = 60,elev = 30)
ax.plot_surface(X,Y,Z, linewidth = 0, edgecolor = "none")
sns.set()
```
